# Incidents

### ChatGPT - Mean Time to Recover

**Best Graph Type:** Line Chart (MTTR over time)

**Why:**

MTTR is also a duration, and you want to show how quickly teams recover from incidents over time.

Line charts allow for:

- Seeing reduction or spikes in MTTR
- Understanding operational maturity
- Correlating with infrastructure or process changes

**Optional secondary views:**

- Scatter Plot of incidents vs. time-to-restore
- Box Plot showing distribution per month
- Bar Chart showing average MTTR per week/month

MTTR often varies widely, so charts that expose distribution are useful.

## Dependencies

```python
import pandas as pd
```

```python
# Useful reference: <https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html>
raw_file_path = '/home/lnx_workspaces/bpp_projects/bpp_module2_project/doraview/data/json/incidents.json'

# Try reading with explicit encoding and error handling
try:
    df_incident = pd.read_json(raw_file_path, encoding='utf-8', convert_dates=["started_at","resolved_at"]).sort_values(by=["started_at"])
    print("\nDataframe info:")
    print(df_incident.info())
    print("\nFirst 5 rows:")
    print(df_incident.head())
except Exception as e:
    print(f"Error: {str(e)}")
```

## Graphs

### Scatter Plot

```python
import plotly.express as px
```

```python
# https://plotly.com/python/line-and-scatter/
fig_lead_scatter = px.scatter(
	data_frame=df_incident,
	title="Mean Time to Recovery Scatter Plot",	# Label for the figure.
	x="started_at",							# Column for use on x-axis
	y="duration_hours",							# Column for use on y-axis
	color="application_id",					# Column for use on color grouping
	trendline="ols",						# Add a trendline
	)

# Apply Plotly colour pallet
fig_lead_scatter.update_layout(template="plotly_dark")

fig_lead_scatter.show()
```
